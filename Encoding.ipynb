```python
# prompt: Generate a code that uses ffmpeg to encode the video to h.264 format and into multiple quality from 1080p to 240p and store it in a folder

import os

def encode_video(input_file, output_dir):
  """Encodes a video to multiple H.264 qualities using FFmpeg.

  Args:
    input_file: Path to the input video file.
    output_dir: Path to the directory where encoded videos will be saved.
  """

  if not os.path.exists(output_dir):
    os.makedirs(output_dir)

  resolutions = [
      (1920, 1080),  # 1080p
      (1280, 720),   # 720p
      (854, 480),    # 480p
      (640, 360),    # 360p
      (426, 240)     # 240p
  ]

  for width, height in resolutions:
    output_file = os.path.join(output_dir, f"encoded_{width}x{height}.mp4")
    command = [
        "ffmpeg",
        "-i", input_file,
        "-vf", f"scale={width}:{height}", # Scale the video to the desired resolution
        "-c:v", "libx264",
        "-crf", "23",
        "-preset", "veryfast",
        "-c:a", "aac",
        "-b:a", "128k",
        output_file
    ]
    !{" ".join(command)} # Execute the ffmpeg command
    print(f"Encoded to {width}x{height} and saved to {output_file}")

# File & Folder location:
input_video = "/content/Recording-12-01-2024.mp4" # Replace with your input video file path
output_directory = "/content/encoded_videos" # Output directory for encoded files
encode_video(input_video, output_directory)
```
